```python

dfZEB=pd.DataFrame()
dfZWB=pd.DataFrame()
dfTD=pd.DataFrame()
dfRY=pd.DataFrame()

dfName=[dfZEB,dfZWB,dfTD,dfRY]
lst=['ZEB.TO','ZWB.TO','TD.TO','RY.TO']

dfAllStocks=pd.DataFrame()
counter=0
for ticker in lst:
    df=pd.DataFrame()
        
    df=yf.download(ticker, start="2017-01-01", end="2017-04-30")
    tic=ticker.replace('.TO','')
    df['Ticker']=tic
    dfAllStocks=pd.concat([dfAllStocks,df],axis=0)
    
    #dfName[counter]=df
    print(dfName[counter])
    print(type(dfName[counter]))
    #counter+=1
    #print(tic)
    #del df
    


```

```python
dfRY=dfAllStocks[dfAllStocks['Ticker']=='RY']

dfRY=dfRY[['Adj Close']].pct_change().dropna()

lst=dfRY['Adj Close']

lst=lst.to_list()

lst


```

```python
import random as r

num_simulations=100
num_datapoints=250

dictSim={}

for i in range(num_simulations):
    
    dictSim[i]=[]
    
    
    for j in range(num_datapoints):
        
        dictSim[i].append(r.choice(lst))
        



dfSim=pd.DataFrame(dictSim)

dfCumReturns=(1 + dfSim).cumprod() - 1
dfCumReturns
        
    
    





```

```python
line_plot = dfCumReturns.plot(legend=False)

```
